```python
def check_password_strength(password):

    #Special characters
    special_characters = "!@#$%^&*()-_=+[]{}|;:,.<>?/`~"
    
    #Initialize flags
    has_uppercase = False
    has_lowercase = False
    has_digit = False
    has_special = False
    is_length_valid = len(password) >= 8
    
    #Check each character in the password
    for char in password:
        if char.isupper():
            has_uppercase = True
        if char.islower():
            has_lowercase = True
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True
    
    #Check if any criteria are not met
    errors = []
    if not is_length_valid:
        errors.append("be at least 8 characters long")
    if not has_uppercase:
        errors.append("contain at least one uppercase letter")
    if not has_lowercase:
        errors.append("contain at least one lowercase letter")
    if not has_special:
        errors.append("contain at least one special character")
    
    #Print the result
    if not errors:
        print("Password is strong")
    else:
        print("Password must " + ", ".join(errors) + ".")

#Ask the user for a password
password = input("Enter a password to check its strength: ")

#Check the password strength
check_password_strength(password)
```

```python
def caesar_cipher(message, key, mode):

    
    result = ""
    
    #Adjust the key for decryption
    if mode == "decrypt":
        key = -key
    
    #Loop through each character in the message
    for char in message:
        #Check if the character is a letter
        if char.isalpha():
            #Determine the base ASCII value for 'A' or 'a'
            base = ord('A') if char.isupper() else ord('a')
            #Shift the character and wrap around using modulo 26
            result += chr((ord(char) - base + key) % 26 + base)
        else:
            #If it's not a letter, leave it unchanged
            result += char
    return result

#Ask the user for inputs
message = input("Enter the message: ")
key = int(input("Enter the key (shift value): "))
mode = input("Enter the mode (encrypt/decrypt): ").lower()

#Perform encryption or decryption
if mode in ["encrypt", "decrypt"]:
    output = caesar_cipher(message, key, mode)
    print(f"Output: {output}")
else:
    print("Invalid mode. Please enter 'encrypt' or 'decrypt'.")
```

```python
import random

def hangman(word_list, max_guesses):

    #Randomly select a word from the list
    word = random.choice(word_list)
    
    #Initialize variables
    guessed_letters = set()  #Letters guessed by the user
    correct_letters = set(word)  #Letters in the word
    incorrect_guesses = 0  #Number of incorrect guesses
    display_word = ["_"] * len(word)  #Display word with underscores

    print("Welcome to Hangman!")
    print(" ".join(display_word))  #Show the initial state of the word

    #Game loop
    while incorrect_guesses < max_guesses:
        guess = input("Guess a letter: ").lower()  #Get user's guess

        #Check if the guess is valid
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single valid letter.")
            continue

        #Check if the letter has already been guessed
        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        #Add the guessed letter to the set of guessed letters
        guessed_letters.add(guess)

        #Check if the guessed letter is in the word
        if guess in correct_letters:
            print("Good guess!")
            #Update the display word with the correctly guessed letter
            for i in range(len(word)):
                if word[i] == guess:
                    display_word[i] = guess
        else:
            print("Incorrect guess.")
            incorrect_guesses += 1

        #Show the current state of the word
        print(" ".join(display_word))

        #Check if the user has guessed the entire word
        if "_" not in display_word:
            print(f"Congratulations, you guessed the word \"{word}\"!")
            return

        #Show remaining guesses
        print(f"You have {max_guesses - incorrect_guesses} guesses left.")

    #If the user runs out of guesses
    print(f"Game over! The word was \"{word}\".")

word_list = ["python", "cobra", "apple", "plumber", "waterfall"]
max_guesses = 6
hangman(word_list, max_guesses)
```
